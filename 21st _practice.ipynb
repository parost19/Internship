```python
# An Exampl using the matplotlib and seaborn for the visulaizatio considering predefined internal dataset.
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')

print(iris.head())

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(iris['sepal_length'], bins=15, color='skyblue', edgecolor='black')
axes[0].set_title('Histogram of Sepal Length')
axes[0].set_xlabel('Sepal Length')
axes[0].set_ylabel('Frequency')

sns.boxplot(x=iris['species'], y=iris['petal_width'], ax=axes[1])
axes[1].set_title('Box plot of Petal Width')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Petal Width')

sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], hue=iris['species'], ax=axes[2])
axes[2].set_title('Scatter plot of Sepal Length vs Sepal Width')
axes[2].set_xlabel('Sepal Length')
axes[2].set_ylabel('Sepal Width')

plt.tight_layout()
plt.show()
```

```python
# taking the numpy and generating the random numbers and using them as the dataset and using the matplot lib for the visualization
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) 
# for reproducibility
data = np.random.normal(loc=0, scale=1, size=100)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(data, bins=15, color='skyblue', edgecolor='black')
axes[0].set_title('Histogram')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

axes[1].boxplot(data, vert=False)
axes[1].set_title('Box plot')
axes[1].set_xlabel('Value')

x = np.random.normal(loc=0, scale=1, size=100)
y = np.random.normal(loc=0, scale=1, size=100)
axes[2].scatter(x, y, color='green')
axes[2].set_title('Scatter plot')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')

plt.tight_layout()

plt.show()
```
